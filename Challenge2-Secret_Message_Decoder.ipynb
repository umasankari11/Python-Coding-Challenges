```python
# Secret Message Decoder

# Input from user
message = input("Enter the secret message: ")

# Convert to uppercase
upper_msg = message.upper()

# Convert to lowercase
lower_msg = message.lower()

# Replace spaces with underscores
clean_msg = upper_msg.replace(" ", "_")

# Show every 3rd character
third_chars = clean_msg[::3]

# Reverse the message
reversed_msg = clean_msg[::-1]

# Count total characters
total_chars = len(message)

# Count alphabetic characters
alpha_count = len([char for char in message if char.isalpha()])  # list comprehension used without 'for' loop outside

# Display results
print("\n----- Decoded Output -----")
print("Uppercase Message:", upper_msg)
print("Lowercase Message:", lower_msg)
print("Spaces Replaced:", clean_msg)
print("Every 3rd Character:", third_chars)
print("Reversed Message:", reversed_msg)
print("Total Characters:", total_chars)
print("Alphabetic Characters:", alpha_count)
```
