**Today's challenge:**

**Topic:** range, for loop, enumerate, list comprehension

**Project Title:** Employee Hours Report

**Scenario:**
 You are analyzing student exam scores to generate performance reports.


**Tasks:**
- Create a list of 10 random exam scores ranging from 0 to 100 using list comprehension.

- Display each score with its index using the enumerate function.

- Find all scores greater than 75 and display them.

- Calculate and display the average score.

```python
# Step 1: Get 10 exam scores from the user, ranging from 0 to 100 using list comprehension

scores = [int(input(f"Enter score {i+1}: ")) for i in range(10)]

# Step 2: Display each score with its index

print("\n---Scores with Index:---")

for index, value in enumerate(scores):

    print("Index", index, ":", value)

# Step 3: Find all scores greater than 75

high_scores = [score for score in scores if score > 75]

print("\n Scores greater than 75:", high_scores)

# Step 4: Calculate average score

average = sum(scores) / len(scores)

print("\nAverage Score:", round(average, 2))
```
