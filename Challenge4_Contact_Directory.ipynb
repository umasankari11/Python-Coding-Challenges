```python
# Contact Directory

# Initial 2D list of contacts
contacts = [
    ["Alice Johnson", "9876543210", "alice@example.com"],
    ["Bob Smith", "9123456780", "bob@example.com"],
    ["Charlie Lee", "9988776655", "charlie@example.com"]
]

# Add a new contact
new_name = input("Enter name: ")
new_phone = input("Enter phone: ")
new_email = input("Enter email: ")
contacts.append([new_name, new_phone, new_email])

# Sort contacts alphabetically by name
contacts.sort(key=lambda x: x[0])

# Display first 3 contacts
print("\n----- First 3 Contacts -----")
for contact in contacts[:3]:
    print(contact)

# Search for a contact by name (partial match)
search_name = input("\nEnter name to search: ").lower()
found_contacts = [c for c in contacts if search_name in c[0].lower()]

print("\n----- Search Results -----")
if found_contacts:
    for c in found_contacts:
        print(c)
else:
    print("No contacts found.")
```
