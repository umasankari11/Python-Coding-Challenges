```python
print("---- Welcome to the Shopping Cart Billing System----")

subtotal = 0
count = 0

while True:
    price = float(input("Enter item price (0 to stop): "))


    # Sentinel value → stop input
    if price == 0:
        break

    # Avoid negative values
    if price < 0:
        print("Negative price not allowed. Try again.")
        continue

    # Valid price → add to cart
    subtotal += price
    count += 1

# Apply discount based on subtotal
if subtotal >= 1000:
    discount_rate = 0.20
elif subtotal >= 500:
    discount_rate = 0.10
elif subtotal >= 200:
    discount_rate = 0.05
else:
    discount_rate = 0.00

discount_amount = subtotal * discount_rate
final_total = subtotal - discount_amount

# Bill Summary
print("\n-------- BILL SUMMARY --------")
print(f"Total items purchased : {count}")
print(f"Subtotal              : ₹{subtotal:.2f}")
print(f"Discount applied      : {discount_rate*100:.0f}%")
print(f"Discount amount       : ₹{discount_amount:.2f}")
print(f"Final amount to pay   : ₹{final_total:.2f}")
print("----------------------------------")
print("Thank you for shopping!")
```
