## üè´ Student Management & Academic Record System

### üéØ Objective:
Understand how to use **Tuples** (immutable data) and **Lists** (mutable, structured data) together  
to manage and organize information efficiently.

---

### üß© Scenario:

A school wants to maintain student data and marks for analysis.

- Each student‚Äôs personal details are **fixed** (Name, Roll Number, Course) ‚Üí stored in a **Tuple**.
- Each student‚Äôs subject marks are **modifiable** ‚Üí stored in a **List**.
- A collection of all students‚Äô marks is stored as a **2D List**.
- For advanced organization, a **3D List** is used to represent marks for multiple batches.

---




```python
#Step 1: Create Tuple for each student (immutable personal info)
student1 = ("San", 101, "BCA")
student2 = ("Uma", 102, "BBA")
student3 = ("Vengi", 103, "BSc")

#Step 2: Create 1D List for marks of each student
stu_mark1=[88,98,100]
stu_mark2=[78,89,80]
stu_mark3=[68,90,90]


#Step 3: Combine all marks into a 2D List
batch1_marks=[stu_mark1,stu_mark2,stu_mark3]

print("Batch 1 Marks:", batch1_marks)


#Step 4: Combine all batches (3D List - two different batches)
batch2_marks = [
    [70, 75, 80],
    [88, 84, 82],
    [91, 93, 89]
]

print("Batch 2 Marks:", batch2_marks)



#Step 5: Display formatted data
all_batches = [batch1_marks, batch2_marks]


print("All Batches Mark")


print(all_batches)

```
