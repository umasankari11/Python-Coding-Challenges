**NumPy Arrays & Operations**

 **Task:**

- Create 1D, 2D, 3D arrays.Perform addition, subtraction, multiplication, division.
- Use functions: mean(), sum(), std(), reshape(), arange(), linspace().
- Find indices of max/min values.

```python
import numpy as np

# 1D Array
arr1 = np.array([1, 2, 3, 4, 5])

# 2D Array
arr2 = np.array([[10, 20, 30],[40, 50, 60]])

# 3D Array
arr3 = np.array([[[1, 2], [3, 4]],
                 [[5, 6], [7, 8]]])

print(f"\n arr1:",arr1)
print(f"\n arr2:",arr2)
print(f"\n arr3:",arr3)


# Perform Addition, Subtraction, Multiplication, Division

print("\n Addition:", arr1 + 2)
print("\n Subtraction:", arr1 - 2)
print("\n Multiplication:", arr1 * 2)
print("\n Division:", arr1 / 2)

#Use functions: mean(), sum(), std(), reshape(), arange(), linspace().

#mean()

print("\n Mean:", np.mean(arr1))

#sum()

print("\n Sum:", np.sum(arr1))

#std()

print("\n Standard Deviation:", np.std(arr1))

# reshape
reshaped = arr1.reshape(1, 5)
print("\n Reshaped array:", reshaped)

# arange
a = np.arange(0, 10, 2)
print("\n Arange:", a)

# linspace
b = np.linspace(1, 10, 5)
print("\n Linspace:", b)

# Find Indices of Max and Min Values

print("\n Index of Max:", np.argmax(arr1))

print("\n Index of Min:", np.argmin(arr1))
```
