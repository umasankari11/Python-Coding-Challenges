```python
# Project: Student Grades & Hobbies Tracker

# Step 1: Create a dictionary with student names and grades
student_grades = {
    "Alice": 85,
    "Bob": 90,
    "Charlie": 78
}

print("Initial student grades:")
for student, grade in student_grades.items():
    print(f"{student}: {grade}")

# Step 2: Allow user to add or update grades for 3 students
for i in range(3):
    name = input("\nEnter student name to add/update grade: ")
    grade = int(input(f"Enter grade for {name}: "))
    student_grades[name] = grade  # Adds new or updates existing
    print(f"Updated grades: {student_grades}")

# Step 3: Create a set of hobbies and allow user to input at least 3 hobbies
hobbies = set()
print("\nEnter at least 3 hobbies:")
while len(hobbies) < 3:
    hobby = input("Enter a hobby: ")
    hobbies.add(hobby)  # Set ensures uniqueness

print(f"\nHobbies set: {hobbies}")

# Step 4: Convert the set of hobbies into a frozenset and display it
permanent_hobbies = frozenset(hobbies)
print(f"Frozen hobbies (immutable): {permanent_hobbies}")
```
