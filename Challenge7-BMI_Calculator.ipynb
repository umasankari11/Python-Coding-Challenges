```python
# Project: BMI Calculator

print("------------BMI Calculator-------------")

print("\n Welcome to the BMI Calculator!")

# Step 1: Ask the user to enter weight and height (integers)
weight = float(input("\n Enter your weight in kilograms (kg): "))
height = float(input("\n Enter your height in meters (m): "))

# Step 2: Calculate BMI
bmi = weight / (height * height)

# Step 3: Classify BMI using conditional statements
if bmi < 18.5:
    category = "Underweight"
elif 18.5 <= bmi < 24.9:
    category = "Normal weight"
elif 25 <= bmi < 29.9:
    category = "Overweight"
else:
    category = "Obese"

# Step 4: Display the result clearly
print(f"\n Your BMI is: {bmi:.2f}")
print(f"\n You are classified as: {category}")
```
